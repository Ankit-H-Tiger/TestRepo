```python
# Importing required packages

import pandas as pd
import numpy as np

import os
import gc
```

```python
path = '/opt/sasdata/dev/PricingEngines/DiscountEngines/Russia/data/raw/gdw_data/'

month_list = os.listdir(path)
month_list.sort()

query_columns = ['sld_menu_itm_id', 'cmbo_pren_sld_menu_itm_id', 'pos_busn_dt']

ans_df = pd.DataFrame()

for month in month_list :
    
    df = pd.read_parquet(path + month + '/data.parquet', engine="pyarrow", columns=query_columns, use_threads=50)
    
    df = df[(df['cmbo_pren_sld_menu_itm_id'] == -1)|(df['cmbo_pren_sld_menu_itm_id'] == df['sld_menu_itm_id'])]
    df['monthid'] = df['pos_busn_dt'].dt.strftime('%Y%m')
    
    ans_df = pd.concat([ans_df, df[['monthid', 'sld_menu_itm_id']].drop_duplicates()], axis=0)
    
    del df
    gc.collect()
    
    print(month)
```
